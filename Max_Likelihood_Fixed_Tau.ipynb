Solve the cubic equation
---------------------
As the second derivative of log(Lx) is negative, the function is concave. The only one real solution is the global maximum.

```python
import math
import sympy as sm
from sympy import *
import numpy as np
from numpy import *
from fractions import Fraction
import matplotlib.pyplot as plt
```

```python
A = Symbol('A', positive=True) # A = n_xxx

B = Symbol('B', positive=True) # B = n_xxy

C = Symbol('C', positive=True) # C = n_xyx

D = Symbol('D', positive=True) # D = n_xyy

h = Symbol('h', nonnegative=True) # h = θ_1

t = Symbol('t', nonnegative=True) # t = τ

d = symbols('d') # d = δ

# the first derivative of the log likelihood function
expr = A*(h + t/h)*(1 + t - d*h - d*t/h)*(1 - t + d*h - d*t/h)*(1 - t - d*h + d*t/h) - B*(h + t/h)*(1 + t + d*h + d*t/h)*(1 - t + d*h - d*t/h)*(1 - t - d*h + d*t/h) + C*(h - t/h)*(1 + t + d*h + d*t/h)*(1 + t - d*h - d*t/h)*(1 - t - d*h + d*t/h) - D*(h - t/h)*(1 + t + d*h + d*t/h)*(1 + t - d*h - d*t/h)*(1 - t + d*h - d*t/h)

sol = solve(expr, d)

for root in sol:
    #print_latex(root)
    print(root)
```

```python
# the first root is the only one real solution
real_solution = sol[0]
```
